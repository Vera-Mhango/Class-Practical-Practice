```python
students = []

def add_student():
    studentID = input("ID number: ")
    name = input("Full Name: ")
    age = int(input("Age: "))

    courses = input("Enter courses (separate with comma): ").split(",")

    student = {
        "id": studentID,
        "name": name,
        "age": age,
        "courses": courses
    }

students.append(student)
print("Student added!\n")

def view_students():
    if not students:
        print("Student not found.\n")
        return

    for student in students:
        print("ID:", student["id"])
        print("Name:", student["name"])
        print("Age:", student["age"])
        print("Courses:", ", ".join(student["courses"]))
        print("-" * 20)

def search_student():
    searchID = input("Enter ID to search: ")

    for student in students:
        if student["id"] == searchID:
            print("Student found: ")
            print(student)
            return

        print("Student Not Found.\n")

def get_unique_courses():
    unique_courses = set()

    for student in students:
        unique_courses.update(student["courses"])

    print("Unique Courses:", unique_courses, "\n")

def remove_student():
    remove_id = input("Enter ID to remove: ")

    for student in students:
        if student["id"] == remove_id:
            students.remove(student)
            print("Student removed!\n")
            return

    print("Student not found.\n")

while True:
    print("1. Add Student")
    print("2. View Students")
    print("3. Search Student")
    print("4. Show Unique Courses")
    print("5. Remove Student")
    print("6. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        add_student()
    elif choice == "2":
        view_students()
    elif choice == "3":
        search_student()
    elif choice == "4":
        get_unique_courses()
    elif choice == "5":
        remove_student()
    elif choice == "6":
        print("Goodbye!")
        break
    else:
        print("Invalid choice\n")
```
